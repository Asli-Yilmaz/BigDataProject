```python
#kategorik değişkneleri incelemek için bar/çubuk grafiği çizmek
def plotBar(degisken,n=20):
  data_tmp=data[degisken]
  data_count=data_tmp.value_counts()
  data_count=data_count[:n]
  plt.figure()
  plt.bar(data_count.index,data_count,color="green")
  ##plt.xticks(data_count.index,data_count.index.values)
  plt.xticks(data_count.index,data_count.index)
  plt.xticks(rotation=75)
  plt.ylabel("Frekans")
  plt.title("Veri Sıklığı - {}".format(degisken))
  plt.show()
  print("{} /n {}".format(degisken,data_count))
```

```python
kategorik_degiskenler=["SiralamaTarihi","Baslik","Tur",
                       "NetflixOzel","YayinTarihi"]
for i in kategorik_degiskenler:
  plotBar(i)
```

##Anomali ve Aykırı Değer Tespiti

Screenshot 2025-03-13 215529.png

```python
def anomaliTespiti(df,ozellik):
  outlier_indices=[]
  for c in ozellik:
    #1.çeyrek
    Q1=np.percentile(df[c],25)
    #3.çeyrek
    Q3=np.percentile(df[c],75)
    #IQR= inter quartile range =çeyrekler açıklığı
    IQR=Q3-Q1
    #AYKIRI değerler için ek adım miktarı
    outlier_step=1.5*IQR
    #AYKIRI değeri ve bulundugu indexi tespit edelim
    outlier_list_col=df[(df[c]<Q1-outlier_step)|(df[c]>Q3+outlier_step )].index
    #tespit edilen indexler burada depolanır
    outlier_indices.extend(outlier_list_col)
  #eşsiz aykırı değerlerin bulunması
  outlier_indices=Counter(outlier_indices)
  #eger bir örnek v adet sürunda farklı bir değer ise bunu aykırı kabul edelim
  multiple_outliers=list(i for i, v in outlier_indices.items() if v>0)
  return multiple_outliers
```

```python
data_anomali=data.loc[anomaliTespiti(data,[ "GunIcindekiSiralama","YilGenelindekiSiralama","KacGundurTop10da"])]
data_anomali
```

```python
plt.figure()
plt.bar(data_anomali.Baslik.value_counts().index,data_anomali.Baslik.value_counts().values)
plt.xticks(rotation=90)
plt.title("Anomaliye rastlanan yapımlar")
plt.ylabel("Frekans")
plt.grid(True,alpha=0.5)
plt.show()
```
