#Regülarizasyon
makine öğrenme algoritmalarında overfitting olmasını engellemeye yarayan bir yöntemdir.

Screenshot 2025-05-06 141018.png

Screenshot 2025-05-06 141246.png

rss'de tüm y(n) karşıya atlıp epsilon değeri de diğer tarafa atıldığınıda B(n)X değerlerinin toplamının 0'a mümkün olduğunca yakın değerler almasını isteriz. rss'in 0 olması demek B(beta) katsayılarını öyle bir hesapladın ki problemin çözümün benim veri setime cuk oturdu demek. Ancak bu durum da overfitting olduğunu gösteriyor olabilir. bu yüzden RSS değerini tek başına 0'a indirgemeye çalışmak yeterli olmayabilir. Özellikle gürültülü bir veri setinde yalnış çıkarımlarda bulunabilir.

Overfitting’i önlemek için modelin kayıp fonksiyonuna ceza terimleri eklenir. Ridge, Lasso ve Elastic Net regülarizasyon yöntemleri bu yaklaşımı kullanır. Ridge, beta katsayılarının büyüklüklerine ceza uygulayarak tümünü küçültmeye çalışır. Lasso ise, bazı katsayıları tamamen sıfırlayarak değişken seçimi yapar. Elastic Net, Ridge ve Lasso’nun avantajlarını birleştirerek hem küçültme hem de sıfırlama etkisi sağlar.

Screenshot 2025-05-06 142213.png

Screenshot 2025-05-06 143139.png

yyukarısaki grafikte kırmızı halkalar rss in tek başına eğittiği veri setimizdir
- ridge yöneteminde sadece iki parametresmiz varsa b(1)^2 + b(2)^2<=s(sabit sayı) grafiği elde ederiz bu da bir dairenin formülüdür. ridge yöneteminde kırmızı halkaların merkezi olan B^ noktasını beta değeri olarak almak yerine bu halkanın bizim daire'miz ile kesiştiği yer olan (B(1),B(2)) noktası bizim beta değerimiz olarak kabul edilir.
- Lasso yönetminde ise iki parametremiz varsa |b(1)|+|b(2)|<=s (sabit) fonksiyonu elde edilir bu da iki boyutlu düzlemde bir karoya denk gelir. burada yine kırmızı halkalar ile gösterilen eğitilmiş verimizin biizm bu grafiğimizle kesiştiği yer beta noktası olarak ela alınır. Burada da (0,b(2)) noktaları beta değeri olarak alınacağından aslında burada b(1) değerinin bizim mevcut problemimizi çözmede hiçbir katkısı olmadığı çıkarımı yapılır(değişken seçimi). bu çıkarımın doğru veya yanlışlığı tartışılabilir.

- bu gibi durumlarda ya elastic net ya da ridge kullanmak daha çok tercih edilir. çünkü lasso bu örnekte olduğu gibi bazı değişkenleri devre dışı bırakabilir.
